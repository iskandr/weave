```python

def fix_tie_up(X_binary, twill_size=8):
    f, p = X_binary.shape
    X_binary = X_binary.copy()
    for i in range(p):
        offset = i % twill_size
        if X_binary.sum(axis=0)[i] < 2:
            X_binary[offset:X_binary.shape[0]:twill_size, i] = True
        if X_binary.sum(axis=0)[i] >= X_binary.shape[0] - 1:
            X_binary[offset:X_binary.shape[0]:twill_size, i] = False
    # though not strictly necessary, also get rid of empty/full rows when possible
    for empty_row_idx in np.where(X_binary.sum(axis=1) == 0)[0]:
        colsums = X_binary.sum(axis=0)
        
        valid_cols = np.where(colsums < (f - 1))[0]
        if len(valid_cols) > 0:
            col_idx = np.random.choice(valid_cols)
            X_binary[empty_row_idx, col_idx] = True
    for full_row_idx in np.where(X_binary.sum(axis=1) == p)[0]:
        colsums = X_binary.sum(axis=0)
        
        valid_cols = np.where(colsums > 1)[0]
        if len(valid_cols) > 0:
            col_idx = np.random.choice(valid_cols)
            X_binary[full_row_idx, col_idx] = False
    return X_binary
```

```python


%matplotlib inline
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.colors
from collections import defaultdict
tieups = defaultdict(dict)

for filename in filenames:
    print("Loading %s" % filename)
    img = Image.open(filename)
    filename  = filename.replace(" ", "-")
    plt.figure(figsize=(12,15))
    
    for i, size in enumerate(sizes):
        print("-- %d" % size)
        thumbnail = img.resize( (size, size), resample=Image.NEAREST)
        parts = filename.split("/")
        dirpath = "/".join(parts[:-1])
        nodir = parts[-1]
        prefix, suffix = nodir.split(".")
        thumbnail_name = prefix + "%d"  % size  + "." + suffix
        thumbnail.save(thumbnail_name)
        tieup = fix_tie_up(np.array(thumbnail)[:, :, 0] > 100)
        Image.fromarray(tieup).save(prefix + "%d"  % size + "-tieup" + "." + suffix)
        plt.subplot(len(sizes), 1, i + 1)
        plt.imshow(tieup, cmap="Greys", interpolation="nearest")
        tieups[filename.split("/")[-1].split(".")[0]][size] = tieup
        
    plt.show()
```

```python
from collections import defaultdict
groupby_size = defaultdict(dict)
for k, d in tieups.items():
    for size, arr in d.items():
        groupby_size[size][k] = arr
js = "{"
for size, d in groupby_size.items():
    if size <= 64:
        js += "\t%d: {\n" % size
        for k, arr in d.items():
            js += "\t\t'%s': boolMatrixFromStrings([\n" % k
            for row in arr:
                js += "\t\t\t'"
                for elt in row:
                    js += "1" if elt else "0"
                js += "',\n"
            js += "\t\t]),\n"
        js += "},\n"
js += "}"
with open('fixed-tie-ups.txt', 'w') as f:
    f.write(js)
    
print(js)
```
